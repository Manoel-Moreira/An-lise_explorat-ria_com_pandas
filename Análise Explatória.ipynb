# Receita:

```python
#Valor total das vendas
df["Valor Venda"].sum()
```

```python
#Valor do custo total:
#Criado uma nova coluna chamada custo, que será o resultado de custo unitário multiplicado por quantidade:
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
#Custo Total:
round(df["Custo"].sum(), 2) #O comando round serve para arredondar, nesse caso o valor em 2 casas decimais
```

# Lucro

```python
#Primeiro passo é criar a coluna lucro
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Total lucro
round(df["Lucro"].sum(), 2)
```

# Envio do pedido

```python
#Criar uma nova coluna para exbir a quantidade de dias que levou para enviar o produto
```

```python
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#Saber a média de dias de envio por marca
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days #Retirado o days para deixar como número inteiro
```

```python
df.head(1)
```

```python
df["Tempo Envio"].dtype
```

```python
df.groupby("Marca")["Tempo Envio"].mean() #Agrupa por marca e traga a média da coluna Tempo Envio
```

#Lucro por ano e por marca

```python
#Primeiro passo é agrupar por ano e por marca e exibir a soma total do lucro
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()

#No minuto 14:30
```
